## Soal 1: Jenis-Jenis Tepi
**Pertanyaan:**
1. Jelaskan empat jenis tepi yang biasa ditemukan dalam citra digital.
2. Berikan contoh masing-masing jenis tepi dalam kehidupan sehari-hari.
3. Mengapa tepi dengan derau memerlukan pendekatan khusus untuk pendeteksiannya?

Jawaban
1. 4 jenis tepi yang biasa di temukan dalam citra yaitu:
  -Tepi Keras (Hard Edge): Tepi ini ditandai dengan perubahan intensitas yang tajam antara dua area yang berbeda. Tepi keras biasanya muncul di antara objek yang memiliki warna atau tekstur yang sangat berbeda.

  -Tepi Lunak (Soft Edge): Tepi lunak menunjukkan perubahan intensitas yang lebih gradual atau halus. Tepi ini sering ditemukan pada objek yang memiliki gradasi warna atau bayangan, seperti pada wajah manusia atau benda-benda yang terkena cahaya.
    
   -Tepi Bertekstur (Textured Edge): Tepi ini muncul di area yang memiliki pola atau tekstur yang kompleks. Tepi bertekstur sering kali sulit untuk dideteksi karena adanya variasi intensitas yang tidak teratur.

   -Tepi Berputar (Curved Edge): Tepi ini menunjukkan perubahan intensitas yang membentuk kurva. Tepi berputar sering ditemukan pada objek bulat atau melengkung, seperti bola atau silinder.

2. - Tepi Keras (Hard Edge)
Contoh: Garis batas antara dua warna cat di dinding. Misalnya, jika satu dinding dicat berwarna biru dan bagian lainnya berwarna putih, tepi antara kedua warna tersebut akan terlihat sangat tajam dan jelas.

  - Tepi Lunak (Soft Edge)
Contoh: Transisi antara kulit wajah dan bayangan di sekitar mata. Ketika seseorang berada di bawah cahaya, peralihan antara area yang terkena cahaya dan area yang lebih gelap di sekitar mata atau pipi akan terlihat lebih halus dan tidak tajam.

  - Tepi Bertekstur (Textured Edge)
Contoh: Permukaan kulit pohon. Ketika melihat kulit pohon, Anda akan melihat pola dan tekstur yang kompleks, di mana tepi antara area yang berbeda (seperti retakan atau tonjolan) tidak teratur dan memiliki variasi intensitas.

   - Tepi Berputar (Curved Edge)
Contoh: Bentuk lingkaran pada piring atau bola. Misalnya, tepi piring yang bulat atau bola basket memiliki kurva yang halus dan tidak memiliki sudut tajam, menciptakan tepi berputar yang jelas.

Contoh-contoh ini menunjukkan bagaimana berbagai jenis tepi dapat ditemukan dalam objek sehari-hari dan bagaimana mereka berkontribusi pada persepsi visual kita.

3. - Gangguan Sinyal Asli
Derau dalam citra digital sering kali muncul sebagai variasi acak dalam intensitas piksel. Hal ini dapat mengaburkan atau mengganggu sinyal asli yang menunjukkan keberadaan tepi. Ketika derau hadir, tepi yang sebenarnya bisa menjadi tidak jelas, dan algoritma deteksi tepi standar mungkin tidak dapat membedakan antara tepi yang valid dan variasi yang disebabkan oleh derau.

   - False Positives dan False Negatives
Derau dapat menyebabkan munculnya "tepi palsu" (false positives) yang tidak ada dalam citra asli, serta kehilangan deteksi tepi yang sebenarnya (false negatives). Ini berarti bahwa algoritma deteksi tepi dapat menghasilkan banyak hasil yang tidak akurat, di mana tepi yang tidak ada terdeteksi sebagai tepi, sementara tepi yang ada tidak terdeteksi.

   - Ketidakstabilan dalam Hasil Deteksi
Derau dapat menyebabkan ketidakstabilan dalam hasil deteksi tepi. Misalnya, jika citra yang sama diproses beberapa kali, hasil deteksi tepi dapat bervariasi tergantung pada tingkat derau yang ada. Hal ini membuat deteksi tepi menjadi tidak konsisten dan sulit untuk diandalkan.

   - Kebutuhan untuk Penyaringan
Untuk mendeteksi tepi yang sebenarnya, sering kali diperlukan teknik penyaringan atau pemrosesan pra-pendeteksian untuk mengurangi derau. Teknik seperti filter Gaussian, filter median, atau metode pemrosesan citra lainnya dapat digunakan untuk menghaluskan citra dan mengurangi pengaruh derau sebelum melakukan deteksi tepi.

   - Penggunaan Algoritma yang Lebih Canggih
Dalam situasi di mana derau sangat signifikan, pendekatan deteksi tepi yang lebih canggih, seperti algoritma berbasis pembelajaran mesin atau metode berbasis model, mungkin diperlukan. Algoritma ini dapat dilatih untuk mengenali pola tepi meskipun ada derau, sehingga meningkatkan akurasi deteksi.


## Soal 2: Pendeteksian Tepi Menggunakan Operator Gradien
**Pertanyaan:**
1. Operator Sobel dan Prewitt digunakan untuk pendeteksian tepi. Pilih salah satu operator tersebut.
2. Implementasikan kode Python menggunakan pustaka seperti NumPy atau OpenCV untuk mendeteksi tepi dalam sebuah citra.
3. Jelaskan langkah-langkah implementasi secara rinci.


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# 1. Membaca gambar
image = cv2.imread('/content/hhhhhh.jpg', cv2.IMREAD_GRAYSCALE)

# 2. Menggunakan operator Sobel untuk mendeteksi tepi
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Deteksi tepi horizontal
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Deteksi tepi vertikal

# 3. Menghitung magnitudo gradien
sobel_magnitude = cv2.magnitude(sobel_x, sobel_y)

# 4. Normalisasi magnitudo ke rentang [0, 255]
sobel_magnitude = cv2.normalize(sobel_magnitude, None, 0, 255, cv2.NORM_MINMAX)

# 5. Menampilkan hasil
plt.figure(figsize=(10, 5))
plt.subplot(1, 3, 1)
plt.title("Original Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.title("Sobel X")
plt.imshow(np.abs(sobel_x), cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.title("Sobel Y")
plt.imshow(np.abs(sobel_y), cmap='gray')
plt.axis('off')

plt.figure(figsize=(5, 5))
plt.title("Sobel Magnitude")
plt.imshow(sobel_magnitude, cmap='gray')
plt.axis('off')

plt.show()

```

Operator Sobel menghitung gradien intensitas piksel dalam dua arah:
Gradien Horizontal (𝐺𝑥G x​) untuk mendeteksi tepi vertikal.
Gradien Vertikal (𝐺𝑦G y​) untuk mendeteksi tepi horizontal.
Hitung gradien dalam kedua arah:
sobel_x = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize=3)  # Gradien horizontal
sobel_y = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize=3)  # Gradien vertikal

Magnitudo gradien (𝐺G) menggabungkan informasi dari kedua arah (𝐺𝑥G x dan 𝐺𝑦G y) menggunakan rumus.image_2024-12-04_192433160.png

## Soal 3: Perbandingan Operator Pendeteksi Tepi
**Pertanyaan:**
1. Berikan perbandingan antara operator Sobel, Prewitt, dan Canny berdasarkan:
   - Akurasi pendeteksian
   - Kompleksitas komputasi
2. Dalam situasi apa operator Canny lebih direkomendasikan dibanding operator lainnya?

jawab

1. berdasarkan Akurasi Pendeteksian:

- Sobel: Akurasi pendeteksian lebih baik dibandingkan Prewitt karena kernel Sobel memiliki bobot yang lebih besar di bagian tengah. Hal ini membuat Sobel lebih sensitif terhadap perubahan intensitas halus dan menghasilkan hasil yang lebih halus.
- Prewitt: Akurasi lebih rendah dibandingkan Sobel karena kernel Prewitt lebih sederhana dan tidak memberikan bobot lebih besar pada piksel di tengah.
- Canny: Akurasi sangat tinggi karena menggunakan langkah-langkah tambahan seperti Gaussian blur untuk mengurangi noise, non-maximum suppression untuk mempertajam tepi, dan thresholding ganda untuk menghindari tepi palsu.

  berdasarkan Kompleksitas Komputasi:

- Sobel: Kompleksitas sedang karena hanya melibatkan operasi konvolusi dengan kernel berbobot.
- Prewitt: Kompleksitas paling rendah karena kernel sederhana tanpa bobot khusus, sehingga lebih cepat tetapi kurang presisi.
- Canny: Kompleksitas tinggi karena melibatkan banyak tahap seperti Gaussian blur, perhitungan gradien, non-maximum suppression, dan thresholding. Hal ini membuatnya lebih lambat tetapi memberikan hasil terbaik.

2. Operator Canny lebih direkomendasikan dalam situasi berikut:

- Kebutuhan Akurasi Tinggi:
  Ketika penting untuk mendeteksi tepi dengan tingkat presisi tinggi dan meminimalkan noise atau tepi palsu.
  Contoh: Pemrosesan gambar medis, seperti deteksi kontur organ dalam citra MRI atau CT scan.
- Lingkungan dengan Noise:
  Jika gambar mengandung noise yang signifikan, Canny menggunakan Gaussian blur untuk mereduksi noise sebelum pendeteksian tepi, sehingga menghasilkan hasil yang lebih bersih dibanding Sobel atau Prewitt.
- Kebutuhan Analisis Lanjut:
  Dalam tugas seperti segmentasi objek, pengenalan pola, atau analisis gambar lainnya yang membutuhkan peta tepi yang detail dan reliabel.
- Ketika Thresholding Penting:
  operator Canny menggunakan thresholding ganda (hysteresis) untuk memastikan bahwa hanya tepi yang signifikan yang dipertahankan, sementara tepi yang lemah disertakan hanya jika terhubung dengan tepi yang kuat.


## Soal 4: Pendeteksian Tepi pada Citra Noisy
**Pertanyaan:**
1. Sebuah citra memiliki derau "salt-and-pepper". Lakukan langkah berikut:
   - Terapkan Gaussian filter untuk mengurangi derau.
   - Gunakan metode Sobel dan Canny untuk mendeteksi tepi.
2. Bandingkan hasil pendeteksian dari kedua metode dengan citra asli.
3. Berikan analisis hasil yang diperoleh.



```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Membaca citra dengan noise salt-and-pepper
image = cv2.imread('/content/uyf.jpg', cv2.IMREAD_GRAYSCALE)

# 1. Gaussian Filter untuk mereduksi noise
blurred_image = cv2.GaussianBlur(image, (5, 5), 1)

# 2. Deteksi tepi menggunakan Sobel
sobel_x = cv2.Sobel(blurred_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(blurred_image, cv2.CV_64F, 0, 1, ksize=3)
sobel_magnitude = cv2.magnitude(sobel_x, sobel_y)
sobel_magnitude = cv2.normalize(sobel_magnitude, None, 0, 255, cv2.NORM_MINMAX)

# 3. Deteksi tepi menggunakan Canny
canny_edges = cv2.Canny(blurred_image, 100, 200)

# Menampilkan hasil
plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.title("Original Image (Salt and Pepper Noise)")
plt.imshow(image, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("Gaussian Filter Applied")
plt.imshow(blurred_image, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 3)
plt.title("Sobel Magnitude")
plt.imshow(sobel_magnitude, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title("Canny Edges")
plt.imshow(canny_edges, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()
```

analisis :
1. sobel edges:
- Gambar yang dihasilkan oleh metode Sobel menunjukkan gradien intensitas dalam dua arah (horizontal dan vertikal).
- Garis tepi utama mobil, seperti kontur bodi, lampu depan, dan atap, terlihat jelas.
- metode Sobel juga menghasilkan beberapa tepi tambahan yang kurang relevan, terutama di area latar belakang.
- Sobel tidak memiliki mekanisme thresholding, sehingga intensitas gradien dalam area berisik (meskipun sudah di-blur oleh Gaussian) masih memberikan kontribusi pada peta tepi.
2. Canny Edges:
- Deteksi tepi menggunakan metode Canny menghasilkan garis tepi yang lebih tajam dan bersih.
- Kontur mobil teridentifikasi dengan baik, terutama pada bagian utama seperti bodi, jendela, dan lampu depan.
- Noise residual hampir sepenuhnya dihilangkan, terutama berkat langkah thresholding ganda dan non-maximum suppression.
- Area latar belakang tidak terlalu terganggu oleh tepi palsu, sehingga fokus utama tetap pada objek mobil.
